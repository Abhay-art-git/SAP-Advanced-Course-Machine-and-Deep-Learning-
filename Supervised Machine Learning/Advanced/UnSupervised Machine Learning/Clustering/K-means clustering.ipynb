```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("clustering.csv")
```

```python
dataset.head()
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(2,21):
    km =KMeans(n_clusters=i)
    km.fit(dataset)
    wcss.append(km.inertia_)
```

```python
plt.figure(figsize=(10,5))
plt.plot([i for i in range(2,21)],wcss , marker = "o")

plt.xticks([i for i in range(2,21)])
plt.grid(axis="x")
plt.show()
```

SELHOUETTE SCORE 

```python
from sklearn.metrics import silhouette_score
```

```python
kmn = KMeans(n_clusters=5)
dataset["predict"]=kmn.fit_predict(dataset)
```

```python
kmn.labels_
```

```python
silhouette_score(dataset , labels=kmn.labels_)
```

```python
dataset
```

```python
sns.pairplot(data = dataset , hue="predict")
plt.show()
```

```python
ss =[]
no_c =[j for j in range(2,21)]
for i in range(2,21):
    km1=KMeans(n_clusters=i)
    km1.fit(dataset)
    ss.append(silhouette_score(dataset,km1.labels_))
```

```python
plt.plot(no_c , ss)
plt.xticks(no_c)
plt.grid(axis="x")
plt.show()
```
