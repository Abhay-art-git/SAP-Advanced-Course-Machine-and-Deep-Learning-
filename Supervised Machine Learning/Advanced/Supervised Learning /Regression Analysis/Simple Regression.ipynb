```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("package.csv")
```

```python
dataset
```

```python
sns.scatterplot(x="cgpa",y="package",data=dataset)
plt.show()
```

```python
x = dataset[["cgpa"]]
y = dataset["package"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
 x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2, random_state=35)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train , y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[1]])
```

```python
y_pred = lr.predict(x)
```

```python
sns.scatterplot(x="cgpa",y="package",data=dataset , label = "original ")
plt.plot(dataset["cgpa"],y_pred , c="red" , label="predict")
plt.legend(["original","predict"])
plt.show()
```
